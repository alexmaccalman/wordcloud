```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('words.csv')
data
# create words list from the input .csv
words = []
for i in range(len(data)):
    word = data.iloc[i,0]
    count = data.iloc[i,1]
    for k in range(count):
        words.append(word)
# create a dataframe and conert to a dictionary
df = pd.DataFrame(words)
freqs = df[0].value_counts().to_dict()
# create the word cloud

wc = WordCloud(width = 1584, height = 400, random_state=20, background_color='black', colormap='Set2', collocations=False).generate_from_frequencies(freqs)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
wc.to_file('wordcloud.png')
```
